**1. Write a Python program to print a right triangle star pattern using nested loops.**

```python
rows = int(input("Enter number of rows: "))
for i in range(1, rows+1):
    for j in range(i):
        print("*", end=" ")
    print()
```

 **Explanation:**
 
First we take the number of rows from the user. The outer loop controls the rows. The inner loop prints stars in each row. end=" " prints stars in the same line. print() moves to the next line.

**2. Write a Python program to print a reverse star pattern using nested loops.**

```python
rows = int(input("Enter number of rows: "))
for i in range(rows, 0, -1):
    for j in range(1, i +1):
        print("*", end=" ")
    print()
```

**Explanation:**

User enters the number of rows. The loop starts from the largest row. Each row prints one star less than the previous row.

**3. Write a Python program to print a number pattern where numbers increase in each row.**

```python
rows = int(input("Enter number of rows: "))
for i in range(1, rows+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

**Explanation:**

The program prints numbers in a pattern. First row prints 1 number. Second row prints 2 numbers. Each row increases the count.

**4. Write a Python program to find the factorial of a given number using a for loop.**

```python
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num+1):
    factorial = factorial * i
print("Factorial is  : ", factorial)
```

**Explanation:**

Factorial means multiplying numbers from 1 to the given number. Example: 5! = 1×2×3×4×5. The loop multiplies numbers step by step.

**6. Write a Python program to check whether a given number is a Prime number or not.**

```python
num = int(input("Enter a number: "))
count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1
if count == 2:
    print("It is a Prime Number")
else:
    print("It is not a Prime Number")
    
```

**Explanation:**

A prime number has exactly two factors. Factors are numbers that divide the number exactly. Example: 7 has factors 1 and 7.

**7. Write a Python program to check whether a given number is a Palindrome number or not.**

```python
num = int(input("Enter a number: "))  # 126
temp = num # stores the 126
reverse = 0  # 1
while num > 0: # 121 > 0 : True
    digit = num % 10 
    reverse = reverse * 10 + digit
    num = num // 10
if temp == reverse:
    print(f"Given {temp} is Palindrome Number")
else:
    print(f"{temp} is not a Palindrome Number")
```

**Explanation:**

Palindrome means same when reversed. Example: 121, 1331. The program reverses the number and compares it with the original.

**8. Write a Python program to check whether a given number is an Armstrong number or not.**

```python
num = int(input("Enter a number: "))   # here are the total main important Armstrong numbers are 153, 370, 371, 407
temp = num
sum_ = 0
while num > 0:
    digit = num % 10
    sum_ = sum_ + (digit ** 3) 
    num = num // 10
if temp == sum_:
    print(f"Given {temp} is Armstrong Number")
else:
    print(f"{temp} is not a Armstrong Number")
```

**Explanation:**

Armstrong number means sum of cubes of digits equals the number. Example: 153 = 1³ + 5³ + 3³ = 153.

**9. Write a Python program to reverse a given string.**

```python
text = input("Enter a string: ")
reverse = ""
for ch in text:
    reverse = ch + reverse
print("Reversed string: ",reverse)
if reverse == text:
    print(f"{text} is a Palindrome")
else:
    print(f"{text} is not a Palindrome")
```

**Explanation:**

The program takes a word from the user. Each character is added in reverse order. Finally the reversed string is printed.

**10. Write a Python program to count the number of vowels in a given string.**

```python
text = input("Enter a string: ").lower()
vowels = ('a', 'e', 'i', 'o', 'u')
count = 0
for ch in text:
    if ch in vowels:
        count += 1
print("Number of vowels:", count)
```

**Explanation:**

Vowels are a, e, i, o, u. The program checks every character in the word. If it is a vowel, the count increases.

**11. Write a Python program to find the largest number in a given list.**

```python
numbers = []
n = int(input("How many numbers do you want to enter? "))
for i in range(1, n+1):
    num = int(input(f"Enter number {i}: "))
    numbers.append(num)
largest = numbers[0]
for i in numbers:
    if i > largest:
        largest = i
print("Largest number is:", largest)
```

**Explanation:**

User enters numbers into a list. The program checks each number. The largest value is stored and printed.

**13. Write a Python program to swap two numbers using a third variable.**

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Before Swapping")
print("a =", a)
print("b =", b)
temp = a
a = b
b = temp
print("After Swapping")
print("a =", a)
print("b =", b)
```

**Explanation:**

First we take two numbers from the user. A third variable called temp is used to store the value of a. Then the value of b is assigned to a. Finally the value stored in temp is assigned to b. This way the values of a and b are exchanged. Example: If a = 5 and b = 10, after swapping a = 10 and b = 5.

**14. Write a Python program to swap two numbers without using a third variable.**

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Before Swapping")
print("a =", a)
print("b =", b)
a = a + b
b = a - b
a = a - b
print("After Swapping")
print("a =", a)
print("b =", b)
```

**Explanation:**

Two numbers are taken from the user. The swapping is done without using any extra variable. First we add both numbers and store the result in a. Then we subtract to find the original value of a and assign it to b. Finally we subtract again to get the original value of b. Example: If a = 4 and b = 7, after swapping a = 7 and b = 4.

**15. Write a Python program to check whether a given year is a Leap Year or not.**

```python
year = int(input("Enter a year: "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```

**Explanation:**

A leap year is a year that has 366 days instead of 365. A year is a leap year if it is divisible by 4, except for years that are divisible by 100. However, years divisible by 400 are leap years. For example, 2000 was a leap year, 1900 was not, and 2024 is a leap year. The program checks these conditions using logical operators.
